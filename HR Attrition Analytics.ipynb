```python
#Attrition Rate by Gender Based
```

```python
# Bar chart for attrition by Gender
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Gender', hue='Attrition', palette='Set2')
plt.title('Attrition Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
#Attrition Rate by Age

# Histogram for Age
plt.figure(figsize=(10, 5))
sns.histplot(df['Age'], kde=True, color='blue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
ax = sns.countplot(data = df, x = 'AgeGroup', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Attrition By Department

# Bar chart for attrition by Department
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Department', hue='Attrition', palette='Set1')
plt.title('Attrition Rate by Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#Attrition Rate By Business Travel
```

```python
# Bar chart for attrition by Business Travel
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='BusinessTravel', hue='Attrition', palette='husl')
plt.title('Attrition Rate by Business Travel')
plt.xlabel('Business Travel')
plt.ylabel('Count')
plt.show()
```

```python
#Attrition Rate By Education Field

# Bar chart for attrition by Education Field
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='EducationField', hue='Attrition', palette='Paired')
plt.title('Attrition Rate by Education Field')
plt.xlabel('Education Field')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#Attrition By Salary/MonthlyIncome

# Box plot for Monthly Income
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['MonthlyIncome'], color='green')
plt.title('Distribution of Monthly Income')
plt.xlabel('Monthly Income')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Selecting relevant numeric features for clustering
features = df[['Age', 'MonthlyIncome', 'JobSatisfaction']]

# Apply KMeans clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(features)

# Visualizing the clusters using scatter plots
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='MonthlyIncome', hue='Cluster', palette='Set1', data=df)
plt.title('Clusters of Employees based on Age and Log Monthly Income')
plt.show()

# Cluster insights
cluster_means = df.groupby('Cluster')[['Age', 'MonthlyIncome', 'JobSatisfaction']].mean()
print("Cluster Centers:")
print(cluster_means)
```

```python
#Overall Attrition Rate

# Calculate the overall attrition rate
attrition_rate = df['Attrition'].value_counts(normalize=True) * 100

# Pie chart to visualize overall attrition rate
plt.figure(figsize=(6, 6))
attrition_rate.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightyellow'])
plt.title('Overall Attrition Rate')
plt.ylabel('')
plt.show()
```

```python
#Conclusion :

#1. The overall attrition rate in this company is 16.2%
#2. Sales Representatives have the highest attrition rate (20.7%), followed by Human Resources (19.04%) and Research Development(13.08%)
#3. Key factors associated with high attrition rates in 3 job roles with the highest attrition rates are job satisfaction, working overtime, and monthly income
#4. Groups that are more likely to leave the company include employees who are single, those in the Human Resources department, male employees, those earning lower monthly income, and employees who feel dissatisfied with their jobs.


```
